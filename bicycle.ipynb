```python

# данные https://raw.githubusercontent.com/jakevdp/data-CDCbirths/master/births.csv


import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display
import datetime
```

```python
df= pd.read_csv('Fremont_Bridge_Bicycle_Counter.csv')
df
```

```python
df1=df.drop_duplicates()
df1.shape
```

```python
df2 = df1.set_index(pd.to_datetime(df1.Date))
df2
```

```python
df2[df2.index.duplicated()]
df2.sort_index(inplace=True)
```

```python
df2.loc['2020-03-08']
```

```python
df3 = df2.truncate(after = '2019-12-31 23:00:00')
df3.loc[df3.isnull().any(axis=1)]
```

```python
df4 = df3.interpolate(method = 'linear', axis = 0)
df4.loc['2013-06-14']
```

```python
# визуализируем все данные по каждому направлению и суммарно
```

```python
df4.plot()
```

```python
# выводим две недели для дальнейшего анализа
```

```python
df4.loc['2015-06-01': '2015-06-14'].plot()
```

```python
# анализируем почасовые тенденции
```

```python
df4.index.hour
```

```python

df_day = df4.groupby(df4.index.time).mean()
hour_tics = df_day.index.values[::4]
hour_tics
g = df_day.plot()
g.set_xticks(hour_tics)
```

```python

df_day = df4.groupby(df4.index.dayofweek).mean()
week_tics = df_day.index.values[::4]
g = df_day.plot()
g.set_xticks(week_tics)
```

```python
day_of_week = ['ПН','ВТ','СР','ЧТ','ПТН','СБ','ВС']
day_type = np.where(df4.index.day_of_week <5, 'workday','weekend')
day_an = df4.groupby([day_type,df4.index.time]).mean()
```

```python
day_an.loc['workday'].plot()
day_an.loc['weekend'].plot()
```

```python
# анализируем почасовые тенденции с разбросом данных используюя sns.lineplot
```

```python
df_sns = df4.copy()
df_sns['day'] = np.where(df4.index.day_of_week <5, 'workday','weekend')
df_sns['time'] = df_sns.index.time
df_sns['month'] = df_sns.index.month
df_sns
sns.set()
sns.lineplot(x = df_sns.time.astype(str), y= df_sns['Fremont Bridge Total'], hue = df_sns.day)
plt.xticks(hour_tics.astype(str))
```
